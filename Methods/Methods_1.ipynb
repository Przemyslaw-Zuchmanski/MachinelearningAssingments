```python
import pandas as pd
import numpy as np
import os

data = pd.DataFrame(
    [
        [1, 92, 6, 4.8, 2060, 49, 1953],
        [2, 70, 6, 5.1, 2660, 36, 1955],
        [3, 58, 6, 6.4, 2530, 49, 1939],
        [4, 99, 2, 6.5, 1740, 28, 1942],
        [5, 55, 8, 2.3, 3520, 77, 1989],
        [6, 76, 4, 5.7, 3750, 55, 1937],
        [7, 62, 6, 4.2, 2720, 43, 1979],
        [8, 92, 6, 6.9, 3130, 39, 1905],
        [9, 71, 5, 4.8, 2100, 54, 1995],
        [10, 70, 6, 4.8, 3340, 29, 1966],
        [11, 77, 4, 7.7, 2430, 53, 1938],
        [12, 79, 4, 5.7, 2700, 47, 1993]
    ],
    columns = ['ID', 'WEI', 'EXE', 'CHO', 'INC', 'HAP', 'BIR']
)

count = data.shape[0]
mean = data.mean(axis = 0)
quantiles = data.quantile([0.25, 0.5, 0.75], axis=0)

print("Count: ", count, "\n")
print("Mean:\n", mean, "\n")
print("Quantiles:\n", quantiles)
print("\n")

for i in range(count):
    if data.values[i][4] < mean[4]:
        print("Person ", i, ": below average income.")
    elif data.values[i][4] > mean[4]:
        print("Person ", i, ": above average income.")
    else:
        print("Person ", i, ": average income.")

print("\n")
kd = pd.read_csv('Anaconda3/kd.csv')
kd.replace('?', np.nan, inplace=True)

kd = kd.astype({"age" : 'float64'})
kd = kd.astype({"bp" : 'float64'})
kd = kd.astype({"bgr" : 'float64'})
kd = kd.astype({"bu" : 'float64'})
kd = kd.astype({"sc" : 'float64'})
kd = kd.astype({"sod" : 'float64'})
kd = kd.astype({"pot" : 'float64'})
kd = kd.astype({"hemo" : 'float64'})

print(kd.min(0, numeric_only=True), "\n")
print(kd.max(0, numeric_only=True), "\n")
print(kd.mean(0, numeric_only=True), "\n")
```
